```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to be integrated
def f(t, y):
    return 10*y - np.sin(2*t*y)

# Set the initial conditions and time range
y0 = np.pi/4
t0 = 0
tmax = 10
dt = 0.00001
t = np.arange(t0, tmax, dt)


# Use the fourth-order Runge-Kutta method to approximate the solution
y = np.zeros_like(t)
y[0] = y0
for i in range(1, len(t)):
    k1 = dt * f(t[i-1], y[i-1])
    k2 = dt * f(t[i-1] + 0.5*dt, y[i-1] + 0.5*k1)
    k3 = dt * f(t[i-1] + 0.5*dt, y[i-1] + 0.5*k2)
    k4 = dt * f(t[i-1] + dt, y[i-1] + k3)
    y[i] = y[i-1] + (1/6) * (k1 + 2*k2 + 2*k3 + k4)

# Plot the solution
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.show()
```
