```python
##CONSTANTS##
k1 = 9e9    #coulombs constant (1/e4pi)
k2 = 1e-7   #magnetic constant (u/4pi)
Qa = 1  #charge on A
Qb = 1 # charge on B
Ma = 1  #mass of a 
Mb = 1  #mass of b
```

```python
a = sphere(radius=0.1, color=color.blue)
b = sphere(radius=0.1, color=color.red)


ra = np.array([1,1,1])
rb =  np.array([11,1,12])
ra = a.pos 
rb= b.pos

a.v =  np.array([0,0,0])
b.v=  np.array([0,0,0])
va= a.v
vb= b.v

a.p = Ma*a.v
b.p= Mb*b.v



dt = 1e-12
numsteps = 1000

def force(ra ,rb ,va,vb,Qa,Qb):
    x = rb -ra 
    fE = k1*Qa*Qb*x/mag(x)**3
    fM = np.cross(va , (k2*Qb*(np.cross(vb,x)))/mag(x**3))
    return fE + fM


for i in range(1,numsteps):
    fa = force(ra,rb ,va,vb,Qa,Qb)
    fb = -fa
    
    a.p= a.p+ fa*dt
    b.p = b.p + fb*dt
    
    a.pos = a.pos + a.p*dt/Ma
    b.pos = b.pos + b.p*dt/Mb
    
```

```python
from vpython import *
##CONSTANTS##
k1 = 9e9    #coulombs constant (1/e4pi)
k2 = 1e-7   #magnetic constant (u/4pi)
Qa = 1.0  #charge on A
Qb = 1.0 # charge on B
Ma = 1.0  #mass of a 
Mb = 1.0  #mass of b

a= sphere(pos=vector(-100,0,0), radius = 10 , color= color.blue)
b= sphere(pos=vector(100,0,0), radius = 10 , color= color.red)

a.v = vector(1.0,0,0)
b.v = vector(1.0,0,0)

a.p = Ma * a.v
b.p = Mb * b.v

t= 0
dt= 0.1
tmax = 1

while t<tmax:
    rBA = b.pos - a.pos
    rAB = -rBA
    fBAe = k1*Qa*Qb*rBA/(mag(rBA))**3
    fBAm = (cross(a.v , k2*Qb*cross(b.v,rBA)/(mag(rBA))**3))
    fBA = fBAe + fBAm
    fAB = -fBA
  
    a.p += fBA*dt
    b.p += fAB*dt
  
    a.pos += a.p*dt/Ma
    b.pos += b.p*dt/Mb
    t+=dt
```

```python
from vpython import *

## CONSTANTS ##
k_e = 9e9        # Coulomb's constant (1/4pi*epsilon_0)
mu_0 = pi * 4e-7 # Magnetic constant (N/A^2)
q_a = 1          # Charge on particle A (C)
q_b = 1          # Charge on particle B (C)
m_a = 1          # Mass of particle A (kg)
m_b = 1          # Mass of particle B (kg)

## INITIAL CONDITIONS ##
a = sphere(pos=vector(0, 0, 0), radius=0.05, color=color.blue)
b = sphere(pos=vector(1, 0, 0), radius=0.05, color=color.red)

a.v = vector(0, 1, 0)
b.v = vector(0, -1, 0)

a.p = m_a * a.v
b.p = m_b * b.v

## SIMULATION PARAMETERS ##
t = 0
dt = 0.01
tmax = 10

## SIMULATION LOOP ##
while t < tmax:
    # Calculate the vector from particle A to particle B
    r_ab = b.pos - a.pos
    r_ba = -r_ab

    # Calculate the electric force on particle A due to particle B
    f_e = k_e * q_a * q_b * r_ab / mag(r_ab)**3

    # Calculate the magnetic force on particle A due to particle B
    v_ab = b.v - a.v
    r_ab_unit = r_ab / mag(r_ab)
    f_b = mu_0 / (4 * pi) * q_a * q_b / mag(r_ab)**2 * cross(v_ab, cross(r_ab_unit, v_ab))

    # Calculate the total force on particle A
    f_a = f_e + f_b

    # Update the momentum and position of particle A
    a.p += f_a * dt
    a.pos += a.p / m_a * dt

    # Calculate the electric and magnetic forces on particle B due to particle A
    f_e = -f_e
    f_b = -f_b

    # Calculate the total force on particle B
    f_b = f_e + f_b

    # Update the momentum and position of particle B
    b.p += f_b * dt
    b.pos += b.p / m_b * dt

    # Update the time
    t += dt
```
